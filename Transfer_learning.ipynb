```python
import pandas as pd
merged_data=pd.read_csv('/home/rgukt/major/dataset/mimic-iv-clinical-database-demo-1.0/core/admissions.csv')
merged_data.isnull().sum()
```

```python
merged_data['admittime'] = pd.to_datetime(merged_data['admittime'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
merged_data['admit_hour'] = merged_data['admittime'].dt.hour          # Extract the hour
merged_data['admit_minute'] = merged_data['admittime'].dt.minute      # Extract the minute
merged_data['admit_second'] = merged_data['admittime'].dt.second      # Extract the second
merged_data['admit_day_of_month'] = merged_data['admittime'].dt.day   # Extract the day of the month
merged_data['admit_month'] = merged_data['admittime'].dt.month        # Extract the month
merged_data['admit_year'] = merged_data['admittime'].dt.year 
merged_data
```

```python
# Convert 'admittime' and 'dischtime' to datetime if they are not already
merged_data['admittime'] = pd.to_datetime(merged_data['admittime'], format='%Y-%m-%d %H:%M:%S')
merged_data['dischtime'] = pd.to_datetime(merged_data['dischtime'], format='%Y-%m-%d %H:%M:%S')

# Calculate the length of stay (los) in hours or days, depending on preference
merged_data['los'] = (merged_data['dischtime'] - merged_data['admittime']).dt.total_seconds() / (60 * 60)  # In hours

# If you prefer los in days, divide by 24
merged_data['los'] = merged_data['los'] / 24  # Converts to d
merged_data
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import joblib
# One-hot encode categorical variables like 'admission_type'
X_upstream = pd.get_dummies(merged_data[['admit_hour', 'admit_minute', 'admit_day_of_month', 'admit_month', 'admit_year', 'admission_type']])

# Define the target variable (length of stay)
y_upstream = merged_data['los']

# Split into training and validation sets
X_train_upstream, X_val_upstream, y_train_upstream, y_val_upstream = train_test_split(X_upstream, y_upstream, test_size=0.2, random_state=42)

# Train a base model on the upstream data (Random Forest Regressor)
base_model = RandomForestRegressor(n_estimators=100, random_state=42)
base_model.fit(X_train_upstream, y_train_upstream)

# Evaluate the base model
y_pred_upstream = base_model.predict(X_val_upstream)
print(f'Mean Squared Error on upstream validation: {mean_squared_error(y_val_upstream, y_pred_upstream)}')
```

```python
joblib.dump(base_model, 'base_model_upstream.pkl')
```

```python
# Merge admissions with ICU stays to create a downstream dataset for ICU patients
admission_details=pd.read_csv('/home/rgukt/major/dataset/mimic-iv-clinical-database-demo-1.0/core/admissions.csv')
icu_details=pd.read_csv('/home/rgukt/major/dataset/mimic-iv-clinical-database-demo-1.0/icu/icustays.csv')
merged_data = pd.merge(admission_details, icu_details, on=['subject_id', 'hadm_id'])

# Assume we want to create a mortality outcome (discharged alive vs. not)
# If discharge_time is None, assume patient died during admission
merged_data['mortality'] = merged_data['dischtime'].isnull().astype(int)

# Now, merged_data contains upstream data (ICU stays and admissions) and a derived downstream outcome (mortality)
merged_data
```

```python
merged_data['admittime'] = pd.to_datetime(merged_data['admittime'])  # Ensure admittime is in datetime format

# Extract useful time features from admittime
merged_data['admit_hour'] = merged_data['admittime'].dt.year
merged_data['admit_day'] = merged_data['admittime'].dt.day
merged_data['admit_month'] = merged_data['admittime'].dt.month
merged_data
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
label_encoder = LabelEncoder()

# Apply label encoding to 'admission_type'
merged_data['admission_type'] = label_encoder.fit_transform(merged_data['admission_type'])
```

```python
merged_data
```

```python
# Load the pre-trained model
base_model = joblib.load('base_model_upstream.pkl')

# Load downstream data (e.g., sepsis patients)
X_downstream = merged_data[[ 'admit_hour', 'admission_type']]  # Example features
y_downstream = merged_data['mortality']  # Target: mortality for sepsis patients

# Split into training and validation sets for downstream task
X_train_downstream, X_val_downstream, y_train_downstream, y_val_downstream = train_test_split(X_downstream, y_downstream, test_size=0.2, random_state=42)

# Fine-tune the model on the downstream dataset
base_model.fit(X_train_downstream, y_train_downstream)

# Evaluate the fine-tuned model
y_pred_downstream = base_model.predict(X_val_downstream)
print(f'Mean Squared Error on downstream validation: {mean_squared_error(y_val_downstream, y_pred_downstream)}')

# Save the fine-tuned model
joblib.dump(base_model, 'fine_tuned_model_downstream.pkl')
```
